In conclusion, this project demonstrates how Exploratory Data Analysis can transform raw data into valuable business insights. ## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Check missing values
print(df.isnull().sum())

# Filling missing values

# Categorical column
df['education_level'].fillna(df['education_level'].mode()[0], inplace=True)

# Numerical columns
df['charging_station_accessibility'].fillna(
    df['charging_station_accessibility'].median(),
    inplace=True
)
df['ev_knowledge_score'].fillna(
    df['ev_knowledge_score'].median(),
    inplace=True
)

# Verify
print("\nMissing values after treatment:")
print(df.isnull().sum())
````
